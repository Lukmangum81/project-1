```python
df_noid = df.iloc[:,2:]
df_noid
```

```python
df_noid.describe()
```

```python
q1= df_noid.quantile(0.25)
q3= df_noid.quantile(0.75)
iqr= q3-q1
print(iqr)
```

```python
df_noid_align, iqr_new = df_noid.align(iqr, axis=1, copy=False, join='outer')
outlier_filter = (df_noid< q1-1.5*iqr_new) | (df_noid > q3+1.5*iqr_new)
outlier_filter
```

```python
# soal nomer 1, 10 pemain dengan goal terbanyak
df_noid[outlier_filter['goals']] \
.loc[0:9,['player_name','goals']]\
.sort_values(by=['goals'],ascending=False)
```
